#### Exercise 3: Working with Strings
**Objective:** Understand how to work with strings in Python.

1. **Concatenate Strings:**
   Concatenate (combine) multiple strings together using the `+` operator:
   ```python
   print("Hello" + " " + "World" + "!")
   ```

2. **Repeat Strings:**
   Repeat a string multiple times using the `*` operator:
   ```python
   print("Hello " * 3)
   ```

3. **Escape Characters:**
   Use escape characters to print special characters like quotes within a string:
   ```python
   print("He said, \"Hello World!\"")
   print('It\'s a beautiful day!')
   ```
